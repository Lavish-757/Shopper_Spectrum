Data Preprocessing

• Objective
- In this notebook, we will clean the raw dataset by removing duplicate records, handling missing values, removing invalid transactions, and creating new features. The cleaned dataset will be used for Exploratory Data Analysis (EDA).

```python
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
```

```python
# Load the dataset

df = pd.read_csv("../data/OnlineRetail.csv", encoding="latin1")
```

```python
# Convert data types

df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["CustomerID"] = df["CustomerID"].astype("Int64")
```

```python
# Dataset shape before cleaning

df.shape
```

• Remove Duplicate Records

- Duplicate rows can affect the accuracy of our analysis. Let's identify and remove them.

```python
# Number of duplicate rows

df.duplicated().sum()
```

```python
# Remove duplicate rows

df = df.drop_duplicates()
```

```python
# Dataset shape after removing duplicates

df.shape
```

• Observation

- Duplicate records have been removed to ensure each transaction appears only once in the dataset.

• Remove Cancelled Transactions

- Invoices starting with 'C' represent cancelled orders. These transactions are removed because they do not represent completed purchases.

```python
# Remove cancelled transactions

df = df[~df["InvoiceNo"].astype(str).str.startswith("C")]
```

```python
# Dataset shape after removing cancelled transactions

df.shape
```

• Observation

- Cancelled transactions have been removed to ensure the analysis is based only on successful purchases.

• Handle Missing Values

- Some records have missing values in important columns. These need to be handled before analysis.

```python
# Check missing values

df.isnull().sum()
```

```python
# Remove rows with missing CustomerID

df = df.dropna(subset=["CustomerID"])
```

```python
# Remove rows with missing Description

df = df.dropna(subset=["Description"])
```

```python
# Check missing values after cleaning

df.isnull().sum()
```

• Observation

- Missing values in 'CustomerID' and 'Description' have been removed, resulting in a cleaner dataset for further analysis.

• Remove Invalid Quantity

- Negative quantities represent returned items, while zero quantities are invalid transactions. These records are removed to keep only valid purchases.

```python
# Keep only positive quantities

df = df[df["Quantity"] > 0]
```

```python
# Check the dataset shape

df.shape
```

• Observation

- Only transactions with a positive quantity are retained for further analysis.

• Remove Invalid Unit Price

- Transactions with a unit price less than or equal to zero are removed because they do not represent valid sales.

```python
# Keep only positive unit prices

df = df[df["UnitPrice"] > 0]
```

```python
# Check the dataset shape

df.shape
```

• Observation

- Only transactions with a valid unit price are retained.

• Clean Text Columns

- Remove extra spaces from text columns to improve data consistency.

```python
# Remove leading and trailing spaces

df["Description"] = df["Description"].str.strip()
df["Country"] = df["Country"].str.strip()
```

• Feature Engineering

- Create a new column named **TotalPrice**, which represents the total amount spent on each transaction.

```python
# Create TotalPrice

df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]
```

```python
# Preview the new column

df[["Quantity", "UnitPrice", "TotalPrice"]].head()
```

• Observation

- The 'TotalPrice' column will be used to analyze revenue, customer spending, and sales performance.

• Final Dataset Overview

- Let's verify that the cleaned dataset is ready for analysis.

```python
# Dataset shape after cleaning

df.shape
```

```python
# Check for remaining missing values

df.isnull().sum()
```

```python
# Display the cleaned dataset

df.head()
```

• Save Cleaned Dataset

- Save the cleaned dataset so it can be used in the next notebook.

```python
# Save the cleaned dataset

df.to_csv("../data/OnlineRetail_cleaned.csv", index=False)
```

• Key Findings

- Duplicate records were removed.
- Cancelled transactions were excluded.
- Missing values in 'CustomerID' and 'Description' were handled.
- Invalid quantities and unit prices were removed.
- A new 'TotalPrice' column was created.
- The cleaned dataset was saved for further analysis.
